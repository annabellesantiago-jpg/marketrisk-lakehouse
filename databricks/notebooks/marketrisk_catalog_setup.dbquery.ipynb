```sql
CREATE CATALOG IF NOT EXISTS market_risk_dev
COMMENT 'MarketRisk Lakehouse — development environment. Contains Bronze, Silver and Gold layers for market risk reporting.';

SHOW CATALOGS;

CREATE SCHEMA IF NOT EXISTS market_risk_dev.bronze
COMMENT 'Raw data exactly as received from the source system - no transformatio applied';

CREATE SCHEMA IF NOT EXISTS market_risk_dev.silver
COMMENT 'Cleaned, deduplicated and enriched data - validated and ready for analytics';

CREATE SCHEMA IF NOT EXISTS market_risk_dev.gold
COMMENT 'Aggregated data ready for reporting. Business metrics - VaR, PnL attribution, limit utilisation and breach flags';

SHOW SCHEMAS IN market_risk_dev;
```
